# Data Discretization

Data Discretization is a pre processing technique used to transform continous or numeric data into bins and categries

```python
import seaborn as sns
import pandas as pd

from sklearn.preprocessing import KBinsDiscretizer

df=sns.load_dataset('titanic')
df.isnull().sum()
```

```python
df.notnull().sum()
```

```python
 # Impute missing values
 
 df['age']=df['age'].fillna(df['age'].median())
df['age'].isnull().sum()
```

```python
# Age discretization

age_disc=KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='uniform')
df['age_bin']=age_disc.fit_transform(df[['age']])

df['age_bin'].value_counts()
```

```python
sns.histplot(df['age_bin'])
```
