# Feature Encoding

Feature encoding is the process of transforming categorical features into numeric. This is necessary because Ml algorithms can nly handle numeric features.

- Label Encoding
- One Hot Encoding
- Ordinal Encoding
- Binary Encoding

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df=sns.load_dataset("tips")
df.head()
```

```python
df['time'].value_counts()
```

```python
# Lets  encode the time with label encoder using sk learn

from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df['encoded_time']=le.fit_transform(df['time'])
df
```

```python
print("Time values before Encoded",df['time'].value_counts())



print("_____________________________________________________")


print("Time values After Encoded",df['encoded_time'].value_counts())
```

```python
# Ordinal encoding of days 

from sklearn.preprocessing import OrdinalEncoder

Oe=OrdinalEncoder(categories=[['Thur','Fri','Sat','Sun']])
df['Encoded_days']=Oe.fit_transform(df[['day']])
```

```python
print("The values before encded",df['day'].value_counts())



print("The days after encoded",df['Encoded_days'].value_counts())
```

```python
# One Hot Encoding on day column
from sklearn.preprocessing import OneHotEncoder
Oh=OneHotEncoder()

Oh.fit_transform(df[['smoker']]).toarray()
```
