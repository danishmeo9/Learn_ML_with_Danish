# **PRINCIPAL COMPONENT ANALYSIS**

`PCA`   is a method used in ML to reduce the dimensionality of the data.The goal of PCA is to identify patterns in data and detect theh correlation between variables.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris=load_iris()

pca=PCA(n_components=2)

iris_pca=pca.fit_transform(iris.data)

plt.scatter(iris_pca[:,0],iris_pca[:,1],c=iris.target)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

# Tips Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler,LabelEncoder


df=sns.load_dataset("tips")


le=LabelEncoder()


# let's make for loop that encoding the categorical columns
for i in df.columns:
    if df[i].dtype=='object' or df[i].dtype=='category':
        df[i]=le.fit_transform(df[i])
    else:
        pass
    
# scalling

scaler=StandardScaler()
scaled_df=scaler.fit_transform(df)

#pca
pca=PCA(n_components=2)

X_pca=pca.fit_transform(scaled_df)

# plot the pca
plt.figure(figsize=(10,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance Explained')
plt.show()



 
```
