# Data Inconsistencies 

####  Inconsistent Format 

- dd-mm-yy
- yyyy-mm-dd
- dd/mm/yy
- 19th May 2019

### Inconsistent Naming Convertion 

- US 
- United States
- United States of America
- USA
- America 

### Typographic errors


- pakistan
- paakistan
- Pakistaan
- PakstaN

### Duplication 

- Duplicated data of same id 
- Same data entered multiple times, sometimes with slight variations.

### Contradictory  


- A person's age is listed as 25 in one record and 32 in another.

- One row says "Gender: Male", another says "Gender: Female" for the same ID
- Age of a son is above than his father 

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data having Inoconsistent Data Types

data={'date':['2021-12-01','12-02-2021','2021/12/03','2021.12.04'],
      'Country':['USA','America','US','United States'],
      'Name':['john Doe','Jane','john doe','Jane Doe'],
      'sales':[100,200,200,None],
      'sales2':[100.0,200.0,300.0,None]}
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
 # Replace the NaN value in the date columns with the date of the first row
 
df['date'].fillna(df['date'].iloc[0], inplace=True)
# Convert the date column to datetime format
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
# Harmonize the country names
country_mapping = {
    'USA': 'United States',
    'America': 'United States',
    'US': 'United States',
    'United States': 'United States'
}

df['Country'] = df['Country'].replace(country_mapping)
df['Country'] = df['Country'].str.title()
df.head()
```

```python
# Correct the typo graphical mistakes in the Name column
name_mapping = {
    'john Doe': 'John Doe',
    'Jane': 'Jane Doe',
    'john doe': 'John Doe',
    'Johne Doe': 'Jane Doe'
}
df['Name'] = df['Name'].replace(name_mapping)
df.head()
```

```python
# Remove duplicates
df=df.drop_duplicates(subset=['date','Country','Name'])
df.head()
```

```python
# Resolving Contridatory Data problems
# sales first column is always greater than sales2 column
df=df.drop(df[df['sales']<=df['sales2']].index)
df.head()
```
