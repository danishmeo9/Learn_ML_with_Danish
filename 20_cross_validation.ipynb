# Cross Validation

Cross -validation is a technique used to evaluate the performance of a model on unseen data. 
In perediction problem a model is usually given a dataset of known dat on which training is run b


The basic idea is to divide the dataset into training and testing sets, train the model on the training set, and then evaluate its performance on the testing set. This process is repeated multiple times, and the average performance is used as an estimate of the model's performance on new data.

1, K-fold cross validation

In `k fold cross validation`, the dataset is divided into `k` parts. The model is trained on `k-1` parts and tested on the remaining part. This process is repeated `k` times, and the average performance is used as an estimate of the model's performance on new data.


## Types f Cross Validation

- K-fold cross validation
- train-test split
- Leave one out cross validation (LOOCV)
- Time series cross validation
- Grou cross validation


### **SUMMARY**
Cross validation is a powerfull technique for evaluating the performance of ML models and can help to ensure thata our models are accurate ,reliable,and able to generalize the new data .

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB


iris = load_iris()

# define the model
model = GaussianNB()

# scores
scores = cross_val_score(model, iris.data, iris.target, cv=5, scoring='accuracy')

# print the scores for each fold and mean score
print("scores for each fold:", scores)
print("mean score:", scores.mean())
print("standard deviation:", scores.std())
```

# K fold cross validation vs train test split

### Diferences

1. Data Usage:
    
> In K-fold Cross-Validation, the data is split into k subsets, and each subset is used as a test set once. The remaining k-1 subsets are used as training sets.

> In traintest split, the data is split into two subsets: a training set and a test set.

2. Computational Cost:

> K-fold cross validation requires more computational resources compared to train-test split, as the model needs to be trained and tested k times.

> Train-test split requires less computational resources, as the model is trained and tested only once. 


`Over all K-Fold Cross-Validation is more efficient than train-test split spcially when the dataset is small and the model is complex. Train test Split is a faster and simpmle techn ique that can be used when the dataset is large and the model is simple.`

```python
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.naive_bayes import GaussianNB
import numpy as np
import pandas as pd
import seaborn as sns


# Let's use tips dataset
df = sns.load_dataset('tips')

X=df[['total_bill','tip','size']]
y=df['sex']

# call the model

model = GaussianNB()

# scores
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')

# print the scores for each fold and mean score
print("scores for each fold:", scores)
print("mean score:", scores.mean())
print("standard deviation:", scores.std())
```
