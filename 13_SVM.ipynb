 # Support Vector Machine

`Support Vector Machines (SVM)`    are powerful supervised learning models mostly for  Classification and Regression and Detecting Outliers.

Hyperplanes define decision boundaries.

Margins measure robustness of separation.

Vectors (w and support vectors) determine the hyperplane’s position and orientation.

Kernels enable linear methods (like SVMs) to handle non-linear data efficiently.
In SVM we take features numeric and target varaiable categorical

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.svm import SVC,SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
# Load data set
df=sns.load_dataset("iris")
df
```

```python
# Let take X and y
X=df.drop("species",axis=1)
y=df["species"]
```

```python
# Split data set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Call the model
model=SVC()
model.fit(X_train,y_train)
```

```python
# prediction
y_pred=model.predict(X_test)
```

```python
# Evaluate the model
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# Draw the confusion matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
